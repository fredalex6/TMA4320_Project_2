### Oppgave 1

#### a) 


Har differensialligningen

$\partial_t \,\phi (x,t) = D \,\partial_x^2 \,\phi (x,t), \hspace{5mm}$ for $t \geq 0 $ og $x \in \mathbb{R}$.

Anvender en fouriertransformasjonen på ligningen får en at

$\partial_t \,\hat{\phi} (w,t) = D(iw)^2 \,\hat{\phi} (w,t) = -Dw^2 \,\hat{\phi} (w,t)$,

som er en separabel differensialligning i frekvensrommet med løsning: $\hspace{5mm} \hat{\phi} (w,t) = C(w) \,e^{-Dw^2 t}$

$C(w)$ får en ved å angi en initialbetingelse $\phi (x,0) = h(x)$, hvilket medfører at $C(w) = \hat{h}(w)$.

Videre kan en definere $\hat{g}(w, t) = e^{-Dw^2 t}$ med inverstransformasjonen $g(x, t) = \frac{1}{\sqrt{2Dt}} e^{-x^2/(4Dt)}$. Løsningen $\phi (x,t)$ blir en konvolusjon mellom $h(x)$ og $g(x, t)$:

$\displaystyle \phi (x,t) = \frac{1}{\sqrt{2\pi}} \left( h(x) * g(x, t) \right) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} h(y) g(x-y, t) \,\mathrm{d}y = \frac{1}{\sqrt{4\pi Dt}} \int_{-\infty}^{\infty} h(y) e^{-(x-y)^2/(4Dt)} \,\mathrm{d}y$

**Vi kan sjekke at løsningen våres stemmer:** 

$\displaystyle \partial_t \phi \sim \partial_t \int_{-\infty}^{\infty} h(y)\,g(x-y,t) \,\mathrm{d}y = \int_{-\infty}^{\infty} h(y) \,\partial_t\,g(x-y, t) \,\mathrm{d}y$

Har at $\partial_t g(x, t) = D \,\partial_x^2 g(x,t)$, slik at

$\displaystyle \partial_t \phi \sim \int_{-\infty}^{\infty} h(y) D\,\partial_x^2\,g(x-y, t) \,\mathrm{d}y = D \partial_x^2 \int_{-\infty}^{\infty} h(y) \,g(x-y, t) \,\mathrm{d}y = D\,\partial_x^2 \,\phi$

Dette er under antagelsen at løsningen og de ulike deriverte konvergerer. Dermed har det som skulle vises blitt vist.

**Fysisk tolkning av h(y):**

Fra utledningen av løsningen $\phi (x,t)$ fant vi at $h(y)$ var gitt som initialbetingelsen $\phi (x,0) = h(x)$. Dette kan også vises ved å ta grenseverdien når $t \rightarrow 0$:

$\displaystyle \lim_{t\rightarrow 0} g(x-y,t) = \lim_{t\rightarrow 0} e^{-(x-y)^2/(4Dt)} = \delta (x-y)$,

da følger det at 

$\displaystyle \phi(x,0) = \int_{-\infty}^{\infty} h(y) \,\delta(x-y) \,\mathrm{d}y = h(x)$.

$h(y)$, eller snarere $h(x)$, kan da tolkes som sannsynlighetstettheten til de ulike partikkelposisjonene ved start.

#### b)


Har at $\displaystyle P(x) = \frac{e^{-\beta V(x)}}{Z}$, der $\displaystyle \sum_{i=0}^{3} e^{-\beta V(x ^{(i)})} = e^{-\beta V(x_0 - 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 + 1)}$. 

$P(x)$ blir da sannsynlighet for at en partikkel fra en vilkårlig posisjon beveger seg mot høyre, venstre eller står i ro. Denne avhenger kun av potensialforskjellen mellom nåværende posisjon og naboposisjoner.


Dermed blir

$\displaystyle p^{-} = \frac{e^{-\beta V(x^{(i)})}}{Z} = \frac{e^{-\beta V(x_0 - 1)}}{e^{-\beta V(x_0 - 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 + 1)}} = \frac{1}{1 + e^{-\beta (V(x_0) - V(x_0-1))} + e^{-\beta (V(x_0+1) - V(x_0-1))}}$

$\displaystyle p^{0} = \frac{e^{-\beta V(x^{(i)})}}{Z} = \frac{e^{-\beta V(x_0)}}{e^{-\beta V(x_0 - 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 + 1)}} = \frac{1}{1 + e^{-\beta (V(x_0-1) - V(x_0))} + e^{-\beta (V(x_0+1) - V(x_0))}}$

$\displaystyle p^{+} = \frac{e^{-\beta V(x^{(i)})}}{Z} = \frac{e^{-\beta V(x_0+1)}}{e^{-\beta V(x_0 - 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 + 1)}} = \frac{1}{1 + e^{-\beta (V(x_0-1) - V(x_0+1))} + e^{-\beta (V(x_0) - V(x_0+1))}}$

#### c)



Når temperaturen $T$ innfrir kravet $k_B T \gg |V(x+1)-V(x)|$ for en vilkårlig $x=x_0$ har en at

$-\beta |V(x_0) - V(x_0-1)| \rightarrow 0$,

$-\beta |V(x_0+1) - V(x_0)| \rightarrow 0$ og følgelig må også

$-\beta |V(x_0+1) - V(x_0-1)| \rightarrow 0$.

Dermed blir $\hspace{5mm} p^- \approx p^0 \approx p^+ \approx 1/3$.

<br>

Når temperaturen $T$ innfrir kravet $k_B T \ll |V(x+1)-V(x)|$ for en vilkårlig $x=x_0$ har en at når $V(x+1) > V(x) > V(x-1)$

$-\beta (V(x_0) - V(x_0-1)) \rightarrow -\infty$ og $-\beta (V(x_0-1) - V(x_0)) \rightarrow \infty$,

$-\beta (V(x_0+1) - V(x_0)) \rightarrow -\infty$ og $-\beta (V(x_0) - V(x_0+1)) \rightarrow \infty$, og følgelig må også

$-\beta (V(x_0+1) - V(x_0-1)) \rightarrow -\infty$ og $-\beta (V(x_0-1) - V(x_0+1)) \rightarrow \infty$.

Dermed blir $\hspace{5mm} p^- \approx 1 \hspace{2mm}$ og $\hspace{2mm} p^0 \approx p^+ \approx 0$.

<br>

Videre når $V(x+1) < V(x) < V(x-1)$ har en at 

$-\beta (V(x_0) - V(x_0-1)) \rightarrow \infty$ og $-\beta (V(x_0-1) - V(x_0)) \rightarrow -\infty$,

$-\beta (V(x_0+1) - V(x_0)) \rightarrow \infty$ og $-\beta (V(x_0) - V(x_0+1)) \rightarrow -\infty$, og følgelig må også

$-\beta (V(x_0+1) - V(x_0-1)) \rightarrow \infty$ og $-\beta (V(x_0-1) - V(x_0+1)) \rightarrow -\infty$.

Dermed blir $\hspace{5mm} p^+ \approx 1 \hspace{2mm}$ og $\hspace{2mm} p^0 \approx p^- \approx 0$.


<br>

I tilfelle der det er et lokalt minima, dvs. $V(x) < V(x \pm 1)$, vil 

$-\beta (V(x_0) - V(x_0-1)) \rightarrow \infty$ og $-\beta (V(x_0-1) - V(x_0)) \rightarrow -\infty$ og

$-\beta (V(x_0+1) - V(x_0)) \rightarrow -\infty$ og $-\beta (V(x_0) - V(x_0+1)) \rightarrow \infty$

Dermed blir $\hspace{5mm} p^0 \approx 1 \hspace{2mm}$ og $\hspace{2mm} p^+ \approx p^- \approx 0$.


<br>

Dersom punktet $x=x_0$ er et lokalt maksima, dvs. $V(x) > V(x \pm 1)$, vil 

$-\beta (V(x_0-1) - V(x_0)) \rightarrow \infty$ og $-\beta (V(x_0+1) - V(x_0)) \rightarrow \infty$.

Dermed blir $p^0 \approx 0$, men det går ikke an å entydig bestemme hva $p^+$ og $p^-$ ettersom vi da må vite differansen $V(x+1) - V(x-1)$ er, som følger fra at

$\displaystyle \frac{p^+}{p^-} \approx e^{-\beta(V(x_0+1) - V(x_0-1))}$.

F.eks. vil tilfellet $V(x+1) = V(x-1)$ medføre at $p^+ \approx p^- \approx 1/2$.


<br>
<br>

Vi kan simplifisere virrevandrings-problemet må å tilnærmere sannsynligetene $p^-$, $p^0$ og $p^+$ med grensebetraktningene over når $\beta k \ll 1$ eller $\beta k\gg 1$.
